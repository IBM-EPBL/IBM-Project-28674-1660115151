```python
#Categorical Unordered Univariate Analysis:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
data.Geography.value_counts(normalize=True)

#plot the bar graph of percentage geography categories
data.Geography.value_counts(normalize=True).plot.barh()
plt.show()
```

```python
#Categorical ordered Univariate Analysis:
data.NumOfProducts.value_counts(normalize=True)

#plot the pie chart of education categories
data.NumOfProducts.value_counts(normalize=True).plot.pie()
plt.show()

```

```python
#Bivariate Analysis
data.plot.scatter(x="EstimatedSalary",y="Balance")
plt.show()
```

```python
#Multi - Variate Analysis
result = pd.pivot_table(data=data, index='Geography', columns='Gender')
print(result)

#create heat map of education vs marital vs response_rate
sns.heatmap(result, annot=True, cmap = 'RdYlGn', center=0.117)
plt.show()
```
